### Instruments

| Instrument | Description |
|-----------|-------------|
| `wti`     | **West Texas Intermediate (WTI) crude oil** – U.S. benchmark crude oil traded on NYMEX |
| `rbob`    | **Reformulated Gasoline Blendstock for Oxygenate Blending (RBOB)** – benchmark for gasoline futures |
| `ho`      | **Heating Oil** – refined petroleum product used for heating and diesel |
| `gasoil`  | **Gasoil** – European diesel futures benchmark, traded on ICE |
| `br`      | **Brent Crude Oil** – international crude oil benchmark, traded on ICE |


### Data Type

#### Prices 

Raw and rolled prices. 

#### Volume 

The number of contracts traded on a given day. Good indicator of market activity.

#### Commitment of Traders (COT)

Gives breakdowns of trader positioning

##### Commercials 

##### Managed Money Accounts 

#### Open Interest

The total number of outstanding contracts (not yet settled). Measures market positioning.

### Availability of Data

| Data Type | Timing | Availability | Notes |
|-----------|--------|--------------|-------|
| **Price** | End of day (settlement) | Same day | Tradable and observable at settlement |
| **Volume** | End of day (settlement) | Same day | Tradable and observable at settlement |
| **Open Interest (OI)** | End of day | **Next morning** | Not available for same-day trading decisions |
| **COT (Commitment of Traders)** | Snapshot as of **Tuesday** | **Published on Friday** | 3-day lag; not tradable in real time |
