# Difference Between Open Interest and Volume in Futures Trading



## Summary Table



| Feature           | **Open Interest (OI)**                              | **Volume**                                      |
|------------------|-----------------------------------------------------|-------------------------------------------------|
| **Definition**    | Total number of **open**, outstanding contracts     | Number of **contracts traded** during the day   |
| **Tracks**        | **Existing positions** still held                   | **New trades** executed                         |
| **Changes when**  | A contract is **opened or closed**                 | A contract is **bought and sold** (always)      |
| **Can Decrease?** | Yes — when positions are closed                     | No — always resets to 0 each day                |
| **Frequency**     | Rolling total, cumulative                          | Daily reset to 0                                |
| **Interpretation**| Measures **commitment / conviction**               | Measures **interest / activity**                |

## Example in WTI Futures

Let’s say today:

- **Trader A** buys 5 contracts from **Trader B** (who is opening a short position)
  - **Volume** increases by **5**
  - **Open Interest** increases by **5** (new positions)

Now later:

- **Trader A** sells 3 contracts to **Trader C** (who is also opening a new long)
  - **Volume** increases by **3**
  - **Open Interest** increases by 3 (Trader C), but Trader A’s are offset, so **no net change**

Finally:

- **Trader B** closes 2 shorts by buying back
  - **Volume** increases by **2**
  - **Open Interest** **decreases by 2**


## Visual Analogy

Think of it like a gym:

- **Volume** = how many people walked in and out **today**
- **Open Interest** = how many people are **currently inside** working out


## Interpretation Use-Cases

| Scenario                     | Meaning                                |
|-----------------------------|----------------------------------------|
| High Volume + Rising OI     | New positions → **trend confirmation** |
| High Volume + Falling OI    | Positions being closed → **reversal?** |
| Low Volume + High OI        | Market is open but **no fresh interest** |


## TL;DR

- **Volume**: Flow — how many contracts traded **today**
- **Open Interest**: Stock — how many contracts are **still open**
- You can have:
  - High volume with no OI change (positions offset)
  - Rising OI = new positions
  - Falling OI = closing positions
