# Why Synthetic Spread?



We model two futures contracts $ F^{(1)} $ and $ F^{(2)} $ as:

$$
F_t^{(1)} = \alpha^{(1)} + \beta^{(1)} M_t + \varepsilon_t^{(1)}
$$
$$
F_t^{(2)} = \alpha^{(2)} + \beta^{(2)} M_t + \varepsilon_t^{(2)}
$$

Then define the **spread** as:

\begin{align}
\text{Spread}_t &= F_t^{(1)} - F_t^{(2)} \\
&= (\alpha^{(1)} - \alpha^{(2)}) + (\beta^{(1)} - \beta^{(2)}) M_t + (\varepsilon_t^{(1)} - \varepsilon_t^{(2)})
\end{align}

- If $ \beta^{(1)} = \beta^{(2)} $, the spread removes the common market factor $ M_t $.
- If $ \beta^{(1)} \neq \beta^{(2)} $, the spread still loads on $M_t$  
   this means the spread contains **systematic risk**.


## Solution: Synthetic Spread

Define the **synthetic spread** as:

$$
\text{Synthetic Spread}_t = F_t^{(1)} - \beta F_t^{(2)}
$$

We aim to choose $ \beta $ such that:

$$
\text{Cov}(\text{Synthetic Spread}_t, M_t) = 0
\quad \text{and} \quad
\text{Cov}(\text{Synthetic Spread}_t, F_t^{(2)}) = 0
$$

By the property of OLS, the residual:

$$
\varepsilon_t = F_t^{(1)} - \beta F_t^{(2)}
$$

is orthogonal to $ F_t^{(2)} $, so:

$$
\text{Cov}(F_t^{(2)}, \varepsilon_t) = 0
\Rightarrow
\text{Cov}(F_t^{(2)}, \text{Synthetic Spread}_t) = 0
$$


## Derivation of Exposure

Letâ€™s expand the synthetic spread:

\begin{align}
\text{Synthetic Spread}_t &= F_t^{(1)} - \beta F_t^{(2)} \\
&= (\alpha^{(1)} + \beta^{(1)} M_t + \varepsilon_t^{(1)}) - \beta (\alpha^{(2)} + \beta^{(2)} M_t + \varepsilon_t^{(2)}) \\
&= (\alpha^{(1)} - \beta \alpha^{(2)}) + (\beta^{(1)} - \beta \beta^{(2)}) M_t + (\varepsilon_t^{(1)} - \beta \varepsilon_t^{(2)})
\end{align}


Thus, the synthetic spread has exposure to $ M_t $ **only if**:

$$
\beta^{(1)} \neq \beta \beta^{(2)} \quad \Leftrightarrow \quad \beta \neq \frac{\beta^{(1)}}{\beta^{(2)}}
$$



## OLS Choice for $ \beta $

Recall from OLS theory:

$$
\beta = \frac{\text{Cov}(F_t^{(1)}, F_t^{(2)})}{\text{Var}(F_t^{(2)})}
$$

Now compute the covariance:

\begin{align}
\text{Cov}(F_t^{(1)}, F_t^{(2)}) &= \text{Cov}(\beta^{(1)} M_t + \varepsilon_t^{(1)}, \beta^{(2)} M_t + \varepsilon_t^{(2)}) \\
&= \beta^{(1)} \beta^{(2)} \text{Var}(M_t)
\quad \text{(assuming } \varepsilon_t^{(1)}, \varepsilon_t^{(2)} \text{ uncorrelated with } M_t \text{ and each other)}
\end{align}
And the variance:
$$
\text{Var}(F_t^{(2)}) = (\beta^{(2)})^2 \text{Var}(M_t) + \text{Var}(\varepsilon_t^{(2)})
$$

So:

$$
\beta = \frac{\beta^{(1)} \beta^{(2)} \text{Var}(M_t)}{(\beta^{(2)})^2 \text{Var}(M_t) + \text{Var}(\varepsilon_t^{(2)})}
$$

## Special Case: Perfect Co-movement

If $ \text{Var}(\varepsilon_t^{(2)}) = 0 $, then:

$$
\beta = \frac{\beta^{(1)}}{\beta^{(2)}}
$$

The synthetic spread **has no market exposure**.


This happens when $ F^{(2)} $ has **perfect co-movement** with $ M_t $ i.e., **no idiosyncratic risk**

But if $ F^{(2)} $ has **idiosyncratic risk**,  then the synthetic spread still retains **some exposure** to $ M_t $.
